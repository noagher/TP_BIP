```python
##compression

import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import dctn, idctn
```

```python
def terme(n, s0, r, i):
    if n == 0:
        return s0
    else:
        return i * (terme(n - 1, s0, r, i) - r)


somme_interet=terme(4,30000,1200,0.01)

print(somme_interet)
```

```python
image=io.imread("squirrel.png")
plt.imshow(image,'gray')
plt.show()
```

```python
DCT=dctn(image,norm='ortho')#the dct transform of the image 
plt.figure(figsize=(10,10))
plt.imshow(np.log(np.abs(DCT[:100,:100])),'gray')#information intressante 
plt.title("Discrete cosine transform")
plt.colorbar()
plt.show()


```

```python
mask_size = 20
mask = np.ones_like(DCT)
mask[mask_size:, mask_size:] = 0
compressed_image = DCT * mask
```

```python
image_compressed=idctn(compressed_image,norm='ortho')

plt.imshow(image_compressed,'gray')
plt.title("compressed image")
plt.show()
```

```python
mse = np.linalg.norm(image- image_compressed)**2 / (image.shape[0] * image.shape[1])

print(f"mean squared error is {mse}")
```

```python
mse_val=[]

for mask_size in range(1,image.shape[0]):
    mask = np.ones_like(DCT)
    mask[mask_size:, mask_size:] = 0
    compressed_image = DCT * mask
    image_compressed=idctn(compressed_image,norm='ortho')
    mse=np.linalg.norm(image- image_compressed)**2 / (image.shape[0] * image.shape[1])
    mse_val.append(mse)


    
```

```python
plt.plot(mse_val)
plt.title("variation of MSE with respect to mask size")
plt.show()
```

```python
##interpolation
from skimage import transform
img = io.imread('chess.png')
angles_90 = [90, 180, 270, 360]  # 360 is equivalent to no rotation
angles_40 = [40*i for i in range(9)]
for angle in angles_90:
    plt.figure(figsize=(8, 4))
    rotated_image=transform.rotate(img,angle)
    mse=np.linalg.norm(img- rotated_image)**2 / (img.shape[0] * img.shape[1])

    plt.subplot(1,2,1)
    
    plt.imshow(img)
    plt.title('original image ')
    plt.subplot(1,2,2)
    plt.imshow(rotated_image)
    plt.title(f"rotation d angle {angle} ")
    plt.show()
    print(f"value of the MSE is {mse}")


```

```python
for angle in angles_40:
    plt.figure(figsize=(8, 4))
    rotated_image_=transform.rotate(img,angle)
    mse_2=np.linalg.norm(img- rotated_image_)**2 / (img.shape[0] * img.shape[1])

    plt.subplot(1,2,1)
    
    plt.imshow(img)
    plt.title('original image ')
    plt.subplot(1,2,2)
    plt.imshow(rotated_image_)
    plt.title(f"rotation d angle {angle} ")
    plt.show()
    print(f"value of the MSE is {mse_2}")
```
