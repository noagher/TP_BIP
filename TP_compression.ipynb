```python
##compression

import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import dctn, idctn
from skimage.morphology import disk
```

```python
image=io.imread("squirrel.png")
plt.imshow(image,'gray')
plt.show()

```

```python
DCT=dctn(image,norm='ortho')

plt.imshow(np.log(np.abs(DCT)),'gray')
plt.title("Discrete cosine transform")
plt.colorbar()
plt.show()


```

```python
mask_size = 20
mask = np.ones_like(DCT)
mask[mask_size:, mask_size:] = 0
compressed_image = DCT * mask
```

```python
image_compressed=idctn(compressed_image,norm='ortho')

plt.imshow(image_compressed,'gray')
plt.title("compressed image")
plt.show()
```

```python
mse = np.linalg.norm(image- image_compressed)**2 / (image.shape[0] * image.shape[1])

print(f"mean squared error is {mse}")
```

```python
mse_val=[]

for mask_size in range(1,image.shape[0]):
    mask = np.ones_like(DCT)
    mask[mask_size:, mask_size:] = 0
    compressed_image = DCT * mask
    image_compressed=idctn(compressed_image,norm='ortho')
    mse=np.linalg.norm(image- image_compressed)**2 / (image.shape[0] * image.shape[1])
    mse_val.append(mse)


    
```

```python
plt.plot(mse_val)
plt.title("variation of MSE with respect to mask size")
plt.show()
```

```python
##interpolation 

from skimage import transform

img=io.imread("chess.png") #charger l image

Rotation_90=[90,180,270,360]#ROTATION MULTIPLE DE 9

for rotation in Rotation_90:
    rotated_image=transform.rotate(img,rotation)
    plt.subplot(1,2,1)
    plt.imshow(rotated_image,'gray')
    plt.title("rotated image")

    plt.subplot(1,2,2)
    plt.imshow(img,'gray')

    plt.title('image original')

    plt.show()
mse_=np.linalg.norm(img- rotated_image)**2 / (img.shape[0] * img.shape[1])
print(f'mse is {mse_}')


```
